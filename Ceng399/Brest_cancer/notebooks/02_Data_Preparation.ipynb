This notebook focuses on preparing the raw radiomics data for machine learning models. We will handle missing values, encode target variables, and scale features as outlined in our project proposal.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib
import os
```

```python
df = pd.read_csv('../data/preprocessed/preprocessed_breast_cancer_data.csv')
```

```python
# Step 5: Train-Test Split
# Separating the dataset into training and testing sets to evaluate the model on unseen data.

# Define x y 
X = df.drop(columns=['target'])
y = df['target']

# Split the data: 80% for training, 20% for testing
# random_state=42 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

# Verify the shapes of the newly created datasets
print(f"Training features (X_train) shape: {X_train.shape}")
print(f"Testing features (X_test) shape: {X_test.shape}")

# Verify that 'stratify' worked correctly by checking the proportions
print("\nTraining Set (y_train) Class Proportions:")
print(y_train.value_counts(normalize=True))

print("\nTesting Set (y_test) Class Proportions:")
print(y_test.value_counts(normalize=True))
```

```python
# Step 6 & 7: Feature Scaling and Exporting Datasets
# Creating and saving both 'unscaled' (for Random Forest) and 'scaled' (for LR and SVM) datasets

os.makedirs('../data/processed', exist_ok=True)


# Saving UNSCALED Data (Ideal for Random Forest)
train_unscaled = pd.concat([X_train, y_train], axis=1)
test_unscaled = pd.concat([X_test, y_test], axis=1)

# Export as CSV
train_unscaled.to_csv('../data/processed/train_unscaled.csv', index=False)
test_unscaled.to_csv('../data/processed/test_unscaled.csv', index=False)

scaler = StandardScaler()

X_train_scaled_arr = scaler.fit_transform(X_train)
X_test_scaled_arr = scaler.transform(X_test)

# Convert numpy arrays back to pandas DataFrames to retain feature names
X_train_scaled = pd.DataFrame(X_train_scaled_arr, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled_arr, columns=X_test.columns, index=X_test.index)

# Part C: Saving SCALED Data (Ideal for Logistic Regression & SVM)
train_scaled = pd.concat([X_train_scaled, y_train], axis=1)
test_scaled = pd.concat([X_test_scaled, y_test], axis=1)


train_scaled.to_csv('../data/processed/train_scaled.csv', index=False)
test_scaled.to_csv('../data/processed/test_scaled.csv', index=False)
```

```python


os.makedirs('../models', exist_ok=True)
joblib.dump(scaler, '../models/standard_scaler.pkl')

print("StandardScaler object saved to '../models/standard_scaler.pkl'.")
```
